```python
import numpy as np
import numexpr
import pandas as pd
```

```python
rng = np.random.RandomState(42)
```

```python
x = rng.rand(int(1E6))
y = rng.rand(int(1E6))
%timeit x+y
```

```python
%timeit np.fromiter((xi+yi for xi,yi in zip(x,y)),dtype=x.dtype,count=len(x))
```

```python
mask = (x>0.5) & (y<0.5)
mask
```

```python
maskexpr = numexpr.evaluate('(x>0.5) & (y<0.5)')
np.allclose(mask,maskexpr)
```

```python
nrows,ncols = 100000,100
rng = np.random.RandomState(42)

df1,df2,df3,df4 = (pd.DataFrame(rng.rand(nrows,ncols)) for i in range(4))
```

```python
df3
```

```python
%timeit df1+df2+df3+df4
```

```python
pd.eval('df1 + df2 + df3 + df4')
```

```python
np.allclose(df1+df2+df3+df4,pd.eval('df1+df2+df3+df4'))
```

```python
df1,df2,df3,df4,df5 = (pd.DataFrame(rng.randint(0,1000,(100,3))) for i in range(5))
```

```python
res1 = -df1*df2/(df3+df4)-df5
res1.head(5)
```

```python
res2 = pd.eval('-df1*df2/(df3+df4)-df5')
res2.head(5)
```

```python
np.allclose(res1,res2)
```

```python
result1 = (df1 <df2) & (df2 <= df3) & (df3!=df4)
result2 = pd.eval('(df1 <df2) & (df2 <= df3) & (df3!=df4)')
np.allclose(result1,result2)
```

```python
res3 = df2.T[0]+df3.loc[1]
res4 = pd.eval('df2.T[0]+df3.loc[1]')
np.allclose(res3,res4)
```

```python
df = pd.DataFrame(rng.rand(1000, 3), columns=['A', 'B', 'C'])

df.head(5)
```

```python
res5 = (df['A'] + df['B'])/(df['C']-1)
res6 = pd.eval('(df.A + df.B)/(df.C - 1)')
np.allclose(res5,res6)
```

```python
res7 = df.eval('(A+B)/(C-1)')
```

```python
np.allclose(res5,res7)
```

```python
df.eval('D = (A+B)/C',inplace=True)
```

```python
df.head(5)
```

```python
df_mean = df.mean(1)
r1 = df['A']+df_mean
r2 = df.eval('A + @df_mean')
np.allclose(r1,r2)
```

```python
resultt1 = df[(df.A<0.5) & (df.B<0.5)]
resultt2 = pd.eval('df[(df.A<0.5) & (df.B<0.5)]')

np.allclose(resultt1,resultt2)
```

```python
resulttt2 = df.query('A<0.5 & B<0.5')
np.allclose(resultt1,resulttt2)
```

```python
Cmean = df['C'].mean()
res1 = df[(df['A']<Cmean) & (df['B']<Cmean)]
res2 = df.query('(A < @Cmean) & (B < @Cmean)')

np.allclose(res1,res2)
```
